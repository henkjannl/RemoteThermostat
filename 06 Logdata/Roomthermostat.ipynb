```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

df = pd.read_csv('Logdata.txt', sep='\t', parse_dates=[['Date', 'Time']], index_col='Date_Time')
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
ax = df[ ['Setpoint', 'measuredTemperature'] ].plot()
hours = mdates.HourLocator(interval = 7*24)
h_fmt = mdates.DateFormatter('%d-%m %H:%M')
ax.xaxis.set_major_locator(hours)
ax.xaxis.set_major_formatter(h_fmt)
plt.show()
```

```python
ax =df[ ['actualBoilerTemperature', 'requestedBoilerTemperature'] ].plot()
x_labels = df.index.strftime('%d-%m %H:%M')
ax.set_xticklabels(x_labels)
plt.show()
```
